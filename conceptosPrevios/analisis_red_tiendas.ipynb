```python
#===========================================
#1️⃣importar la libreria de pandas
#===========================================
import pandas as pd
import matplotlib.pyplot as plt
#===========================================
#2️⃣ cargar los archivos CVS
#===========================================
#usamos pd.read_cvs
ventas_df =pd.read_csv("ventas.csv")
inventarios_df=pd.read_csv("inventarios.csv")
satisfaccion_df =pd.read_csv("satisfaccion.csv")
#==========================================
#3️⃣Mostrar las primeras filas 
#==========================================
print("===ventas---")
print(ventas_df.head(),"\n")
print("===Inventarios---")
print(inventarios_df.head(),"\n")
print("===Satisfaccion---")
print(satisfaccion_df.head(),"\n")
#==========================================
#4️⃣Limpiar datos (elimina filas con valores nulos)
#==========================================
ventas_df=ventas_df.dropna()
inventarios_df=inventarios_df.dropna()
satisfaccion_df=satisfaccion_df.dropna()
#==========================================
#5️⃣Analisis de ventas totales 
#==========================================
ventas_df["Ventas_Totales"]=ventas_df["Cantidad_Vendida"]*ventas_df["Precio_Unitario"]
print(ventas_df)
ventas_totales=ventas_df.groupby("ID_Tienda")["Ventas_Totales"].sum().reset_index()
print(ventas_totales)
#==========================================
#6️⃣Rotacion de inventario
#==========================================
ventas_inventarios=pd.merge(ventas_df,inventarios_df,on=["ID_Tienda","Producto"],how="inner")
#==========================================
# merge combina las filas de ambos DF
# on=["ID_Tienda","Producto"]#clave de union es obligatoria 
# how="inner" solo conserva las filas que tienen coincidencia en ambos DF(interseccion)
#==========================================
print("========================ventas de Inventarios===========================")
print(ventas_inventarios)
ventas_inventarios["Rotacion_Inventario"]=ventas_inventarios["Cantidad_Vendida"]/ventas_inventarios["Stock_Disponible"]
print(ventas_inventarios)

#==========================================
#7️⃣Analizar la satisfacción del cliente y filtrar las tiendas con una satisfacción menor al 60%.

# 1. Calcular ventas totales y stock total por tienda
ventas_totales = ventas_df.groupby("ID_Tienda")["Cantidad_Vendida"].sum().reset_index()
stock_total = inventarios_df.groupby("ID_Tienda")["Stock_Disponible"].sum().reset_index()

# 2. Unir ventas y stock
ventas_stock = pd.merge(ventas_totales, stock_total, on="ID_Tienda", how="inner")

# 3. Calcular rotación (ventas / stock)
ventas_stock["Rotacion"] = ventas_stock["Cantidad_Vendida"] / ventas_stock["Stock_Disponible"]

# 4. Filtrar tiendas con rotación < 0.10 (10 %)
tiendas_criticas = ventas_stock[ventas_stock["Rotacion"] < 0.10]

# 5. Mostrar resultados
print("\n=== Tiendas con inventarios críticos (<10% rotación) ===")
print(tiendas_criticas)


#==========================================
#  Filtrar tiendas con satisfacción menor al 60 %
#==========================================
"""
# 1. Calcular el promedio de satisfacción por tienda
promedio_satisfaccion = satisfaccion_df.groupby("ID_Tienda")["Satisfaccion_Promedio"].mean().reset_index()

# 2. Filtrar las que tengan promedio < 60
tiendas_baja_satisfaccion = promedio_satisfaccion[promedio_satisfaccion["Satisfaccion_Promedio"] < 0.6]

# 3. Mostrar resultados
print("\n=== Tiendas con satisfacción menor al 60% ===")
print(tiendas_baja_satisfaccion)
"""
#==========================================
#  Calcular la mediana de las ventas totales
#==========================================

# 1. Sumar las ventas por tienda
ventas_totales = ventas_df.groupby("ID_Tienda")["Cantidad_Vendida"].sum().reset_index()

# 2. Calcular la mediana
mediana_ventas = ventas_totales["Cantidad_Vendida"].median()

# 3. Mostrar resultado
print(f"\n📌 La mediana de las ventas totales es: {mediana_ventas}")

#==========================================
#  Calcular la desviación estándar de las ventas
#==========================================

# 1. Sumar las ventas por tienda
ventas_totales = ventas_df.groupby("ID_Tienda")["Cantidad_Vendida"].sum().reset_index()

# 2. Calcular la desviación estándar
desviacion_ventas = ventas_totales["Cantidad_Vendida"].std()

# 3. Mostrar resultado
print(f"\n📌 La desviación estándar de las ventas es: {desviacion_ventas}")




import numpy as np

#==========================================
#  Simulación de proyecciones de ventas
#==========================================

# 1. Obtener ventas históricas por tienda
ventas_totales = ventas_df.groupby("ID_Tienda")["Cantidad_Vendida"].sum().reset_index()

# 2. Definir parámetros para la simulación
np.random.seed(42)  # 🔒 Para resultados reproducibles
crecimiento_promedio = 0.05   # 5% de crecimiento anual esperado
desviacion_crecimiento = 0.02 # Variación del 2%

# 3. Simular 12 meses futuros para cada tienda
proyecciones = {}

for _, fila in ventas_totales.iterrows():
    ventas_actuales = fila["Cantidad_Vendida"]
    # Generar factores de crecimiento aleatorios para 12 meses
    factores = np.random.normal(crecimiento_promedio, desviacion_crecimiento, 12)
    # Calcular ventas proyectadas
    ventas_futuras = ventas_actuales * (1 + factores)
    proyecciones[fila["ID_Tienda"]] = ventas_futuras

# 4. Mostrar proyecciones
print("\n=== Proyecciones de ventas futuras ===")
for tienda, ventas in proyecciones.items():
    print(f"Tienda {tienda}: {np.round(ventas, 2)}")


#generar datos  aleatorieos en el tiempo
tiempo=np.arange(100)
ventas_aleatorias=np.random.rand(100)*1000
print(ventas_aleatorias)
#proyectar las ventas con una tendencia lineal
pendiente=50
interseccion=1000
ventas_proyectadas=pendiente*tiempo+interseccion+ventas_aleatorias
#visualizar los resultados 

plt.plot(tiempo, ventas_proyectadas, label="ventasproyectadas")
plt.xlabel("Tiempo")
plt.ylabel("ventas proyectadas")
plt.legend()
plt.show ()
```
